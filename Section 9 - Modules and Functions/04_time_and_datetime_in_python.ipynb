```python
import time
from time import time as my_timer
import random

input("Please Enter to Start:---")

wait_time = random.randint(1,6)
time.sleep(wait_time)

start_time = my_timer()

input("Please Enter to Stop:---")

end_time = my_timer()

print("Started at" + time.strftime("%X", time.localtime(start_time)))
print("Ended at" + time.strftime("%X", time.localtime(end_time)))

print("Your Reaction time was {} seconds".format(end_time - start_time))
```

```python
import time

print("The epoch on this system starts at" + time.strftime("%c", time.gmtime(0)))

print("The Current TimeZone is {0} with an offset of {1}".format(time.tzname[0], time.timezone))

if time.daylight != 0:
    
    print("\tDaylight Saving Time is in effect for this Location")
    print("\tThe DST timezone is" + time.tzname[1])
    
else:
    
    print("Not Applicable in the Current Location.")
    
print("Local time is " + time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
print("UTC time is " + time.strftime("%Y-%m-%d %H:%M:%S", time.gmtime()))
```

**MORE ON TIMEZONES**

```python
import datetime
import pytz

local_time = datetime.datetime.now()
print("Naive local time {}".format(local_time))

utc_time = datetime.datetime.utcnow()
print("Naive UTC time {}".format(utc_time))
```

```python
aware_local_time = pytz.utc.localize(local_time)
aware_utc_time = pytz.utc.localize(utc_time)

print("Aware Local time is : --", aware_local_time)
print("Aware UTC time is : --", aware_utc_time)
```

```python
aware_local_time = pytz.utc.localize(local_time).astimezone()
print("Aware local time {}, timezone {}".format(aware_local_time, aware_local_time.tzinfo))

aware_utc_time = pytz.utc.localize(utc_time)
print("Aware UTC {}, timezone {}".format(aware_utc_time, aware_utc_time.tzinfo))
```

```python
dt1 = pytz.utc.localize(datetime.datetime.fromtimestamp(s)).astimezone(gb)
dt2 = pytz.utc.localize(datetime.datetime.fromtimestamp(t)).astimezone(gb)

print("{} seconds since the epoch is {}".format(s,dt1))
print("{} seconds since the epoch is {}".format(t,dt2))
```

```python
import datetime
import pytz

available_zones = {
    "1" : "Africa/Abidjan",
    "2" : "US/Indiana-Starke",
    "3" : "Pacific/Palau",
    "4" : "Mexico/BajaNorte",
    "5" : "Europe/Minsk",
    "6" : "Australia/Queensland",
    "7" : "Asia/Samarkand",
    "8" : "America/Inuvik",
    "9" : "Zulu"
}

print("Please choose a timezone from the list below (or 0 to quit)")

for place in sorted(available_zones):
    print("\t{}  :  {}".format(place, available_zones[place]))


while True:
    
    choice = input()
    
    if choice == "0":
        break
        
    if choice in available_zones.keys():
        tz_to_display = pytz.timezone(available_zones[choice])
        world_time = datetime.datetime.now(tz=tz_to_display)
        print("The time in {} is {}, and timezone is {}".format(available_zones[choice], world_time.strftime("%A %x %X %Z"), world_time.tzname()))
        print("Local time is {}".format(datetime.datetime.now().strftime("%A %x %X")))
        print("UTC time is {}".format(datetime.datetime.utcnow().strftime("%A %x %X")))
```
