### PARSING JSON DATA

```python
import json

json_data_source = "temperature_anomaly.json"

with open(json_data_source, encoding = "utf-8") as data:
    anomalies = json.load(data)
    
print(anomalies)
```

```python
print(anomalies["description"])
```

```python
print(anomalies["data"])
```

```python
print(anomalies["citation"])
```

```python
for key, value in anomalies["description"].items():
    print(f" {key} :  {value}")
```

```python
for key, value in anomalies["data"].items():
    print(f" {key} :  {value}")
```

```python
for key, value in anomalies["citation"].items():
    print(f" {key} :  {value}")
```

Citation Contains String Objects.

```python
for year, value in anomalies["data"].items():
    year, value = int(year), float(value)
    
    print(f"{year} ....{value : 6.2f}")
```

### PARSING DATA FROM THE INTERNET

```python
import json
import urllib.request

json_data_link = "https://www.ncei.noaa.gov/access/monitoring/climate-at-a-glance/global/time-series/globe/land_ocean/1/11/1880-2022.json"

with urllib.request.urlopen(json_data_link) as json_stream:
    
    data = json_stream.read().decode("utf-8")
    
    anomalies = json.loads(data)
```

```python
print(anomalies)
```

```python
print(anomalies["data"])
```

```python
for year, value in anomalies["data"].items():
    
    year, value = int(year), float(value)
    
    print(f"{year} ...{value : 6.2f}")
```
