```python
import csv

input_filename = 'country_info.txt'

with open(input_filename, encoding='utf-8', newline='') as countries_data:
    
    sample = ""
    for line in range(4):
        sample = sample + countries_data.readline()
    
    sample = countries_data.read()                                   ## read() will read the entire data to a string.
    country_dialect = csv.Sniffer().sniff(sample)
    country_dialect.skipinitialspace = True
    
    countries_data.seek(0)                             ## seek method is used to point the pointer to strat from the begonning.
    
    #print(sample)
    
    country_reader = csv.reader(countries_data, dialect = country_dialect)      ## this time reader will start from the beginning stage.        
    for row in country_reader:
        print(row)

        
        ## here what is happening.
        ## sample variable will be the entire data read by read() method
        ## then we introduced the seek method to satrt reading the data from the beginning.
        ## so basically the entire data is been read twice completely.
        
        ## in fact, to check the data, we dont need to read whole data to sample variable.
        ## wwe should only make to read two or three lines to check.
        
        
attributes = [
    "delimiter",
    "doublequote",
    "escapechar",
    "lineterminator",
    "quotechar",
    "quoting",
    "skipinitialspace"
]

for attribute in attributes:
    print(f"{attribute} : {repr(getattr(country_dialect, attribute))}")
```
