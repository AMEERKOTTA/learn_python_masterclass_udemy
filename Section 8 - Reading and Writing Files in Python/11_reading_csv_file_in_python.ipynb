### READING CSV FILE IN PYTHON

+ CSV :- Comma Seperated Values.

```python
import csv

csv_filename = "OlympicMedals_2020.csv"

with open(csv_filename, encoding = "utf-8", newline="") as csv_file:
    reader = csv.reader(csv_file)
    for row in reader:
        print(row)
```

In the above Code, We have Column Headers as another Row.
+ To understand and print Column Headers Seperately.
+ Use the readline() method.
+ Read the first line using readline() method.
+ Then reader will start reading from the Second Line.

```python
import csv

csv_filename = "OlympicMedals_2020.csv"

with open(csv_filename, encoding = "utf-8", newline="") as csv_file:
    
    ## add code here
    headers = csv_file.readline().strip("\n").split(",")  ## reading the first line using readline() function
    print(f"The Column Headers are {headers}")            ## then the reader will only start reading from second line.
    
    reader = csv.reader(csv_file)                         ## this way we will be able to get the column name seperately.
    for row in reader:
        print(row)
```

### Quoting in a CSV file.

+ In the first cell Output, look at the numerical values.
+ They are quoted by single quotes.
+ That is, all numenrical values have been read as Strings.

+ We can change that by introducing one more agrument in the reader function.
+ Add quoting = csv.QUOTE_NONNUMERIC.
+ Inside reader Function.
+ Check the Second Cell from here.

+ NOw thw numeric values are not included in the String.
+ Now we have floats as numeric values.

+ This method only works if there are all numeric values are quoted.
+ That is one limitation.

```python
import csv

csv_filename = "cereal_grains.csv"

with open(csv_filename, encoding = "utf-8", newline = "") as csv_file:
    
    reader = csv.reader(csv_file)
    
    for row in reader:
        
        print(row)
```

```python
with open(csv_filename, encoding = "utf-8", newline = "") as csv_file:
    
    reader = csv.reader(csv_file, quoting = csv.QUOTE_NONNUMERIC)
    
    for row in reader:
        
        print(row)
```
