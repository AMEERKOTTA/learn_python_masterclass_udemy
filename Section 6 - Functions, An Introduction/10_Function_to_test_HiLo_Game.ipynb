```python
def guess_binary(answer, low, high):
    """
    Function yo test the Hi Lo game.
    Here in this function is used to test the scenorio for every number from 1 to 1000
    So, as a result for every number from 1 to 1000, we will get the number guesses to make that number guessed correctly
    
    :param answer :-- Answer is the Number that we are going to guess.
    :param low :-- low is 1
    :param high :- high is 1000
    :return :--The function will return the number which is guessed and in how many guess we made it correctly.
    """
    
    ## initiate the variable guesses to some value
    guesses = 1
    
    ## implement a while loop to check the Hi Lo situation.
    while True:
        #print("\tGuessing in the range {0} to {1}".format(low,high))
        
        ## initiate guess variable
        guess = low + (high - low) // 2
        #print("The Guessed Value:--", guess)
        
        ## there will be 4 situations in guessing game.
        ## if guess < answer, we have to guess higher.
        ## if guess > answer, we have to guess lower.
        ## if guess = answer, then guess is correct.
        ## if guess is an invalid number, then we have to guess again.
        
        if guess < answer:
            #print("Guessed value is Less than Answer, Please Guess higher")
            low = guess + 1
            
        elif guess > answer:
            #print("Guessed value is Higher than Answer, Please Guess Lower")
            high = guess - 1
        
        elif guess == answer:
            return guesses
        
        guesses += 1
        
```

```python
low = 1
high = 1000

for number in range(low,high+1):
    number_of_guesses = guess_binary(number,low,high)
    print("{0} guessed in {1}".format(number,number_of_guesses))
```

**Counting the COrrect Guesses**

+ Calculate the maximum Number of Guesses.
+ Also count how many times that maximum number is returned.

```python
correct_count = 0
max_guesses = 0

for number in range(low, high +1):
    number_of_guesses = guess_binary(number,low,high)
    print("{} guessed in {}".format(number,number_of_guesses))
    
    if number_of_guesses > max_guesses:
        max_guesses,correct_count = number_of_guesses,1
        
    elif number_of_guesses == max_guesses:
        correct_count += 1 
        
print("I guessed without being told {} times. Maximum {} guesses".format(correct_count,max_guesses))
```
